```python
import seaborn as sns
```

```python
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
tips = sns.load_dataset('tips')
```

```python
flights = sns.load_dataset('flights')
```

```python
tips.head()
```

```python
#plot-1 dist plot
sns.displot(tips['total_bill'],kde=False)
```

```python
sns.displot(tips['total_bill'],kde=False,bins=30)
```

```python
#plot-2 joint plot
sns.jointplot(x='total_bill',y='tip',data=tips)
```

```python
sns.jointplot(x='total_bill',y='tip',data=tips,kind='hex')
```

```python
sns.jointplot(x='total_bill',y='tip',data=tips,kind='reg')
```

```python
sns.jointplot(x='total_bill',y='tip',data=tips,kind='kde')
```

```python
#plot-3 pairplot
sns.pairplot(tips)
```

```python
sns.pairplot(tips,hue='sex')
```

```python
sns.pairplot(tips,hue='sex',palette='coolwarm')
```

```python
#plot-4
sns.rugplot(tips['total_bill'])
```

```python
#kde = kernel density estimation plots
sns.rugplot(tips['total_bill'])
```

```python
import numpy as np
```

```python
#the most basic categorical plot: bar plot
sns.barplot(x='sex', y='total_bill', data= tips, estimator=np.std)
```

```python
sns.barplot(x='sex', y='total_bill', data= tips)
```

```python
#countplot is basically a barplot but the estimator is explicitly count, y is already chosen
sns.countplot(x='sex', data=tips)
```

```python
sns.boxplot(x='day',y='total_bill',data=tips)
```

```python
sns.boxplot(x='day',y='total_bill',data=tips, hue='smoker')
```

```python
#hue is a huge advantage of seaborn
sns.violinplot(x='day',y='total_bill',data=tips)
```

```python
sns.violinplot(x='day',y='total_bill',data=tips, hue='smoker')
```

```python
sns.violinplot(x='day',y='total_bill',data=tips, hue='smoker', split='true')
```

```python
sns.stripplot(x='day',y='total_bill', data=tips, jitter=False)
#problem with stripplot: does not show overlapping points
```

```python
sns.stripplot(x='day',y='total_bill', data=tips, jitter=True)
```

```python
sns.stripplot(x='day',y='total_bill', data=tips, jitter=True, hue='smoker')
```

```python
#swarmplot is a combination of a violinplot and a split stripplot 
sns.swarmplot(x='day', y='total_bill', data=tips)
#go way too wide, not recommended for large datasets
```

```python
sns.violinplot(x='day',y='total_bill',data=tips)
sns.swarmplot(x='day', y='total_bill', data=tips, color='red')
```

```python
#factorplot is the most general plot
#kind point is the characteristic of factorplots
#factorplot was replaced by caatplot in the new version of seaborn
sns.catplot(x='day', y='total_bill', data=tips, kind='bar')
```

```python
iris = sns.load_dataset('iris')
iris.head()
```

```python
iris['species'].unique()
```

```python
sns.pairplot(iris)
```

```python
g = sns.PairGrid(iris)
g.map(plt.scatter)
```

```python
g = sns.PairGrid(iris)
g.map_diag(sns.histplot)
g.map_upper(plt.scatter)
g.map_lower(sns.kdeplot)
```
