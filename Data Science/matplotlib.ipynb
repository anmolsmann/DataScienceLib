```python
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
import numpy as np
x = np.linspace(0,5,11)
y = x**2
```

```python
x
```

```python
y
```

```python
plt.plot(x,y)
```

```python
plt.plot(x,y,'r-')
```

```python
plt.plot(x,y)
plt.xlabel('X')
plt.ylabel('Y')
plt.title('New')
```

```python
plt.subplot(1,2,1)
plt.plot(x,y,'r')

plt.subplot(1,2,2)
plt.plot(y,x,'b')
```

```python
#object oriented method
```

```python
fig = plt.figure()
axes = fig.add_axes([0.1,0.1,0.8,0.8])
axes.plot(x,y)
axes.set_xlabel('X')
axes.set_ylabel('Y')
axes.set_title('Title')
```

```python
fig = plt.figure()
axes1 = fig.add_axes([0.1,0.1,0.8,0.8])
axes2 = fig.add_axes([0.2,0.3,0.3,0.3])
axes1.plot(x,y)
axes2.plot(y,x)
```

```python
fig,axes = plt.subplots(1,2)
axes[0].plot(x,y)
axes[1].plot(y,x)
```

```python
fig,axes = plt.subplots(nrows=2,ncols=1,figsize=(8,2))
axes[0].plot(x,y)
axes[1].plot(y,x)
plt.tight_layout()
```

```python
#saving a figure
```

```python
fig.savefig('firstMatPLotLib')
```

```python
fig = plt.figure()
axes = fig.add_axes([0,0,0.5,0.5])
axes.plot(x,x**2,label='X^2')
axes.plot(x,x**3,label='X^3')
axes.legend(loc=0)
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,x**2,'green',lw=2,alpha=0.5,marker='o')
ax.plot(x,x**3,'black',lw=1,linestyle='--',marker='*')
```
